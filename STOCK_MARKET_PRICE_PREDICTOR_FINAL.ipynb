STOCK PRICE PREDICTOR USING LSTM (Long Short-term memory)

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler 
import math
```

Extracting 80% of stock data to be used as the training set. 

```python
values=y.values
print(len(values))
training_data_length=math.ceil(len(values)*0.8)
print(training_data_length)
```

```python
scaler=MinMaxScaler(feature_range=(0,1))                #to normalize and reshape data
scaled_data=scaler.fit_transform(values.reshape(-1,1))

train_data_lstm=scaled_data[0: training_data_length, :]
x_train_lstm=[]
y_train_lstm=[]

for i in range(90, len(train_data_lstm)):
    x_train_lstm.append(train_data_lstm[i-90:i, 0])#creating 90-days window of historical prices (i-90) as feature data (x_train)
    y_train_lstm.append(train_data_lstm[i, 0])     #following 90-days window as label data (y_train)
    
x_train_lstm=np.array(x_train_lstm)  #converting since numpy array is data format accepted by tensorflow
y_train_lstm=np.array(y_train_lstm)
```

```python
test_data_lstm=scaled_data[training_data_length-90: , : ]
x_test_lstm=[]
y_test_lstm=values[training_data_length:]

for i in range(90, len(test_data_lstm)):
  x_test_lstm.append(test_data_lstm[i-90:i, 0])

x_test_lstm=np.array(x_test_lstm)
x_test_lstm=np.reshape(x_test_lstm, (x_test_lstm.shape[0], x_test_lstm.shape[1], 1))
```

```python
from tensorflow.keras.layers import Dense, Dropout, LSTM
```

```python
model=keras.Sequential()
model.add(layers.LSTM(100, return_sequences=True, input_shape=(x_train_lstm.shape[1], 1))) #LSTM layer with 100 network units
model.add(Dropout(0.2)) # to prevent overfitting
model.add(layers.LSTM(100, return_sequences=False)) #will only return last output in the sequence
model.add(Dropout(0.2)) # to prevent overfitting
model.add(layers.Dense(25)) #densely connected neural network layer
model.add(layers.Dense(1))
model.summary() #summary of LSTM network architecture
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train_lstm, y_train_lstm, batch_size= 1, epochs=3)
```

Finding RMSE value to examine model performance as it will help calculate degree of discrepancy between predicted and actual prices

```python
y_pred_lstm=model.predict(x_test_lstm)
y_pred_lstm=scaler.inverse_transform(y_pred_lstm) #to denormalize the predicted stock prices
rmse=np.sqrt(np.mean((y_pred_lstm - y_test_lstm)**2))
print("RMSE=", rmse)
```

Plot of Predicted vs Actual Stock prices

```python
lstm_data=data.filter(['Close'])
train=lstm_data[:training_data_length]
validation=lstm_data[training_data_length:]
validation['Predictions'] = y_pred_lstm
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date')
plt.ylabel('Close Price USD ($)')
plt.plot(train)
plt.plot(validation[['Close', 'Predictions']])
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```

This shows that the LSTM model is very effective and better than Linear Regression to predict stock prices.
